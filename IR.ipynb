```python
import pandas as pd
import xlrd
import re
import nltk
from nltk.corpus import stopwords

path='Nepal_train.xlsx'
dataframe = pd.read_excel(path, usecols=[1])
mylist=dataframe["text"].tolist()
lis=[]
for line in mylist:
    lis.append(line.lower())

for i in range(len(lis)):
    lis[i] = re.sub(r'http\S+', '', lis[i])
    lis[i]=' '.join(word for word in lis[i].split(' ') if not word.startswith('@'))
    lis[i]=re.sub(r'\W+', ' ', lis[i])
    lis[i]=' '.join([word for word in lis[i].split() if word not in (stopwords.words('english'))])
    lis[i]=re.sub(r'[0-9\.]+', '', lis[i])
```
